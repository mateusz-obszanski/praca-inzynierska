```python
from pathlib import Path
import json
```

```python
from bin.rand_exp_config import generate_rand_conf
from bin.refactored_experiment import experiment_tsp
from libs.data_loading.utils import ExperimentType
from libs.data_loading.loaders import get_env_data
```

```python
env_data = get_env_data(Path("data/environments/map_2021-12-15_08-57-55_tsp.json"))
conf = generate_rand_conf(
    ExperimentType.TSP,
    env_data,
    population_size=10,
    generation_n=100,
    timeout=60,
    early_stop_iters=50
)
```

```python
conf_path = Path("test_experiment_config.json")
conf.save_to_json(conf_path)
```

```python
results_path = "test_experiment_results.json"
experiment_tsp(
    str(conf_path),
    results_path,
    population_size=10,
    generation_n=100,
    exp_timeout=60,
    early_stop_n=50,
)
with open(results_path, "r") as f:
    d = json.loads(f.read())
    print("\n".join(f"{k}: {v}" for k, v in d.items()))
```
